```python
import csv

with open('0506.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()  # 파일의 모든 행을 리스트로 읽어옵니다.

data = []  # 출력할 데이터를 저장할 리스트를 초기화합니다.

for line in lines:
    line = line.replace('  ', ' ').strip()  # '  ' (두 개의 공백)을 ' ' (한 개의 공백)으로 변경하고, 양쪽 공백을 제거합니다.
    words = line.split(' ')  # ' ' (한 개의 공백)을 기준으로 단어를 분리하여 리스트로 저장합니다.
    if len(words) >= 2:  # 단어가 2개 이상인 경우에만 출력합니다.
        data.append([words[0], words[1], words[2], words[3], words[-2]])  # 두 번째 단어와 맨 뒤에서 첫 번째 단어를 리스트로 추가합니다.

with open('0506.csv', 'w', encoding='cp949', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['달', '일', '시간','장비','번호판'])  # 헤더 행을 쓰기합니다.
    writer.writerows(data)  # 데이터 행을 쓰기합니다.
```

```python
with open("0506.txt", "r", encoding='UTF8') as f:
    # 파일을 읽기 모드로 열고, with 구문으로 자동으로 파일을 닫습니다.
    for line in f:
        # 파일에서 한 행씩 읽어옵니다.
        line = line.replace('  ', ' ')
        words = line.split(' ')  # ' ' (공백)을 기준으로 단어를 분리하여 리스트로 저장

        print(words[0], words[1], words[2], words[3], words[-2])  # 두 번째 단어와 맨 뒤에서 첫 번째 단어 출력
```
